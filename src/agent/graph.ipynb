```python
import uuid

from dotenv import load_dotenv
from IPython.display import Image, display
from langgraph.checkpoint.memory import MemorySaver

from configuration import MY_REPORT_STRUCTURE
from graph import report_workflow
```

```python
load_dotenv()
```

```python
graph = report_workflow.compile(checkpointer=MemorySaver())
display(Image(graph.get_graph(xray=1).draw_mermaid_png()))
```

```python

job_description = """
Full job description
Support Platform Specialist, gTech Users and Products at Google

Minimum qualifications:
Bachelor's degree in Engineering, Computer Science, a related field, or equivalent practical experience.
3 years of experience in coding with one or more programming languages (e.g., Java, C/C++, Python).
3 years of experience in working with web technologies (e.g., HTML, CSS, JavaScript, or HTTP).
Experience in technical troubleshooting, and managing internal/external partners or customers.
Preferred qualifications:
Master's degree in Engineering, Computer Science, Business, or a related field.
3 years of experience in online operations.
Experience working with database technologies (e.g., SQL, NoSQL).
Ability to work globally and cross-functionally exhibiting strong leadership, project management, communication, and teamwork.
Ability to take initiative, multi-task and complete projects on-time
Excellent technical and functional skills in a changing cross-platform environment.
"""

thread = {"configurable": {"thread_id": str(uuid.uuid4()),
                           "search_api": "tavily",
                           "planner_provider": "openai",
                           "planner_model": "o3-mini",
                           "writer_provider": "openai",
                           "writer_model": "gpt-4o-mini",
                           "max_search_depth": 2,
                           "number_of_queries": 5,
                           "top_k": 2,
                           "report_structure": MY_REPORT_STRUCTURE,
                           }}
```

```python
async for event in graph.astream({"topic":job_description}, thread, stream_mode="updates"):
    print(event)
    print("\n")
```

```python
from IPython.display import Markdown
final_state = graph.get_state(thread)
report = final_state.values.get('final_report')
Markdown(report)
```
